```python
import sys
from PIL import Image
import numpy as np
import skimage.io as io
import skimage.color as color
from skimage import transform
import matplotlib.pyplot as plt
import cv2
```

```python
def RGBtoHSV(img):
    img[:][:] = img[:][:]/255.0

    # defining the following terms for convenience
#     Cmax = max(R_dash, G_dash, B_dash)
#     Cmin = min(R_dash, G_dash, B_dash)
#     delta = Cmax - Cmin
#     # hue calculation
#     if (delta == 0):
#       H = 0
#     elif (Cmax == R_dash):
#       H = (60 * (((G_dash  - B_dash) / delta) % 6))
#     elif (Cmax == G_dash):
#       H = (60 * (((B_dash  - R_dash) / delta) + 2))
#     elif (Cmax == B_dash):
#       H = (60 * (((R_dash  - G_dash) / delta) + 4))
    for i in range(len(img)): 
        for j in range( len(img[i]) ):
            (b,g,r) = img[i][j]
            val = max(b,g,r)
            c = val - min(b,g,r)

            sat = -1
            if val == 0:
                sat = 0
            else:
                sat = c/val

            hue = -1
            if c == 0:
                hue = 0
            elif val == r:
                hue = (60 * ((g-b)/c)) % 6
            elif val == g:
                hue = (60 * ((b-r)/c) + 2)
            elif val == b:
                hue = (60 * ((r-g)/c) + 4)
                
            img[i][j] = (hue,sat,val)
            
            # Creating array back into image
            data = Image.fromarray(img,'RGB')
            data.save("puppy.png")
            
    return img
```
